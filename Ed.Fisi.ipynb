```python
from datetime import datetime
import math

def ponto_virgula(mensagem = ''):
    while True:
        entrada = str(input(mensagem).replace(',', '.')).strip()
        if entrada.isalpha():
            print(f'\033[0;31m"{entrada}" não é um valor válido, tente novamente\033[m')
        elif entrada == '':
            return 'Não Informado'
        else:
            num = float(entrada)
            return num
```

```python
Cadastro_Aluno = {}
Cadastro_Aluno['Nome'] = input('Digite o Nome do Aluno: ').title()

def calcular_idade(cadastro):
    while True:
        try:
            data_nascimento = input('Digite Dia/Mês/Ano: ')
            if data_nascimento == '':
                cadastro['Idade'] = 'Não informado'
                return cadastro
            data_nascimento = datetime.strptime(data_nascimento, "%d/%m/%Y")
            hoje = datetime.today()
            idade = hoje.year - data_nascimento.year
            if (hoje.month, hoje.day) < (data_nascimento.month, data_nascimento.day):
                idade -= 1
            cadastro['Idade'] = idade
            # Formatando a data para apenas Dia/Mês/Ano
            cadastro['Data_de_nascimento'] = data_nascimento.strftime("%d/%m/%Y")
            return cadastro
        except ValueError:
            print("\033[0;31mData inválida! Use o formato Dia/Mês/Ano.\033[m")

Cadastro_Aluno = calcular_idade(Cadastro_Aluno)

```

```python
def Sexo(cadastro):
    while True:
        sexo = input('Digite o Sexo do Aluno: ').title()[0]
        if sexo == 'M':
            cadastro['Sexo'] = 'Masculino'
            break
        elif sexo == 'F':
            cadastro['Sexo'] = 'Feminino'
            break
        else:
            print('tente novamente')
    return cadastro  

Cadastro_Aluno = Sexo(Cadastro_Aluno)
```

```python

def Percentual_de_Gordura(cadastro):
    
    Sete_Dobras = {}
    sexo = cadastro.get('Sexo', '').title()
    idade = cadastro.get('Idade', 0)
    Peso_total = cadastro.get('Peso')
    sete_dobras = cadastro.get('Medidas_Dobras', {})
    soma_dobras = sete_dobras.get('Soma Total', 0)

    if not isinstance(idade, (int, float)) or idade <= 0:
            cadastro['Percentual_de_Gordura'] = "Idade inválida ou não definida."
            return cadastro
    if soma_dobras == 0:
        raise ValueError("A soma das sete dobras não foi calculada corretamente.")
    if Peso_total is None or Peso_total <= 0:
        raise ValueError("Peso total inválido ou não informado.")
    
    if sexo == 'Masculino':
        Densidade_Corporal = 1.112 - 0.00043499 * soma_dobras + 0.00000055 * (soma_dobras**2) - 0.00028826 * idade
        percentual = (495 / Densidade_Corporal) - 450
    elif sexo == 'Feminino':
        Densidade_Corporal = 1.097 - 0.00043499 * soma_dobras + 0.00000055 * (soma_dobras**2) - 0.00028826 * idade
        percentual = (495 / Densidade_Corporal) - 450
    else:
        raise ValueError("Sexo inválido ou não definido.")
    
    Percent_Gordura = round(percentual, 2)
    Massa_Magra = Peso_total - (Peso_total * Percent_Gordura / 100)
    Massa_Gorda = Peso_total - Massa_Magra
    cadastro['Sete_Dobras'] = {
         'Percentual_de_Gordura' : Percent_Gordura,
        'Massa_Gorda' : round(Massa_Gorda, 2),
        'Massa_Magra' : round(Massa_Magra, 2)
    }
    cadastro['Sete_Dobras'] = Sete_Dobras
    return cadastro

#Cadastro_Aluno = Percentual_Residual(Cadastro_Aluno)
#Cadastro_Aluno = Massa_Gorda(Cadastro_Aluno)

```

```python
def Duas_Dobras(cadastro):
    """
    Calcula a densidade corporal para crianças e adolescentes (9 a 16 anos),
    com base nos protocolos atualizados de McArdle.
    """
    dobras = cadastro.get('Medidas_Dobras', {})
    sexo = cadastro.get('Sexo', '').title()
    idade = cadastro.get('Idade', 0)
    
    if not dobras or not idade:
        raise ValueError("Dados insuficientes para o cálculo. Certifique-se de que todas as medidas foram informadas.")
        
    if sexo == 'Masculino':
        if 9 <= idade <= 12:
            densidade = 1.108 - 0.027 * math.log10(dobras['Tricipital']) - 0.038 * math.log10(dobras['Subescapular'])
        elif 13 <= idade <= 16:
            densidade = 1.130 - 0.055 * math.log10(dobras['Tricipital']) - 0.026 * math.log10(dobras['Subescapular'])
        else:
            densidade = "Idade fora do intervalo suportado (9 a 16 anos)."
    elif sexo == 'Feminino':
        if 9 <= idade <= 12:
            densidade = 1.088 - 0.014 * math.log10(dobras['Tricipital']) - 0.036 * math.log10(dobras['Subescapular'])
        elif 13 <= idade <= 16:
            densidade = 1.114 - 0.031 * math.log10(dobras['Tricipital']) - 0.041 * math.log10(dobras['Subescapular'])
        else:
            densidade = "Idade fora do intervalo suportado (9 a 16 anos)."
    else:
        raise ValueError("Sexo inválido ou não informado.")
    
    Duasdobras = {}

    if isinstance(densidade, (int, float)):
        percentual_gordura = (4.95 / densidade - 4.50) * 100
        Duasdobras['Percentual_de_Gordura'] = round(percentual_gordura, 2)
        Duasdobras['Densidade_Corporal'] = round(densidade, 4)
    else:
        Duasdobras['Percentual_de_Gordura'] = "Idade fora do intervalo suportado (9 a 16 anos)."
        Duasdobras['Densidade_Corporal'] = densidade

    cadastro['Duas_Dobras'] = Duasdobras
    return cadastro


Cadastro_Aluno = Duas_Dobras(Cadastro_Aluno)


```

```python
def Tres_dobras_Guedes(cadastro):
    """
    Calcula a densidade corporal com base nas três dobras cutâneas
    para homens e mulheres.
    """
    sexo = cadastro.get('Sexo', '').title()
    dobras = cadastro.get('Medidas_Dobras', {})

    try:
        if sexo == 'Masculino':
            dobras_necessarias = ['Tricipital', 'Supra_Iliaca', 'Abdominal']
        elif sexo == 'Feminino':
            dobras_necessarias = ['Coxa', 'Supra_Iliaca', 'Subescapular']
        else:
            raise ValueError("Sexo inválido ou não definido.")


        if not all(dobra in dobras for dobra in dobras_necessarias):
            raise KeyError("Falta de dados necessários para o cálculo.")

        # Calcula a densidade corporal
        soma_dobras = sum(dobras[dobra] for dobra in dobras_necessarias)

        if sexo == 'Masculino':
            densidade = 1.17136 - 0.06706 * math.log10(soma_dobras)
        elif sexo == 'Feminino':
            densidade = 1.16650 - 0.07063 * math.log10(soma_dobras)

        cadastro['Densidade_Corporal'] = round(densidade, 2)
        return cadastro

    except KeyError as e:
        cadastro['Densidade_Corporal'] = ValueError(f"Erro: {e}")
        return cadastro
    except Exception as e:
        cadastro['Densidade_Corporal'] = ValueError(f"Erro ao calcular densidade: {e}")
        return cadastro

    

Cadastro_Aluno = Tres_dobras_Guedes(Cadastro_Aluno)
```

```python
def Cinco_Dobras(cadastro):
    """
    Calcula a densidade corporal e percentual de gordura com base nas cinco dobras cutâneas.
    """
    dobras_cutaneas = cadastro.get('Medidas_Dobras', {})
    sexo = cadastro.get('Sexo', '').title()
    idade = cadastro.get('Idade', 0)
    densidade_corporal = None

    try:
        if sexo == 'Masculino' and 18 <= idade <= 61:
            dobras_requeridas = ['Peitoral', 'Abdominal', 'Coxa']
            if not all(dobra in dobras_cutaneas for dobra in dobras_requeridas):
                raise KeyError("Faltam dobras necessárias para o cálculo.")
            
            soma_dobras = sum(dobras_cutaneas[dobra] for dobra in dobras_requeridas)
            densidade_corporal = (
                1.1093800
                - 0.0008267 * soma_dobras
                + 0.0000016 * (soma_dobras ** 2)
                - 0.0002574 * idade
            )

        elif sexo == 'Feminino' and 18 <= idade <= 55:
            dobras_requeridas = ['Tricipital', 'Supra_Iliaca', 'Coxa']
            if not all(dobra in dobras_cutaneas for dobra in dobras_requeridas):
                raise KeyError("Faltam dobras necessárias para o cálculo.")
            
            soma_dobras = sum(dobras_cutaneas[dobra] for dobra in dobras_requeridas)
            densidade_corporal = (
                1.0994921
                - 0.0009929 * soma_dobras
                + 0.0000023 * (soma_dobras ** 2)
                - 0.0001392 * idade
            )
        else:
            raise ValueError("Faltam dados cadastrais ou o aluno não está na faixa etária requerida.")

        # Calcula o percentual de gordura usando a fórmula de Siri
        percentual_gordura = round((4.95 / densidade_corporal - 4.50) * 100, 2)

    except KeyError as e:
        cadastro['Cinco_Dobras'] = {"Erro": f"Falta a medida: {e}"}
        return cadastro
    except Exception as e:
        cadastro['Cinco_Dobras'] = {"Erro": f"{e}"}
        return cadastro

    cadastro['Cinco_Dobras'] = {
        'Densidade_Corporal': round(densidade_corporal, 4),
        'Percentual_de_Gordura': percentual_gordura,
    }
    return cadastro


Cadastro_Aluno = Cinco_Dobras(Cadastro_Aluno)
```

```python
def Seis_Dobras(cadastro):
    """
    Calcula o percentual de gordura com base em seis dobras cutâneas:
    Subescapular, Tríceps, Coxa, Supra-Ilíaca, Abdome e Peitoral.
    
    Fórmula: G% = (S6) x 0.095 + 3.64
    Onde S6 é a somatória das dobras cutâneas indicadas.
    """
    dobras_cutaneas = cadastro.get('Medidas_Dobras', {})
    dobras_requeridas = ['Subescapular', 'Tricipital', 'Coxa', 'Supra_Iliaca', 'Abdominal', 'Peitoral']
    
    try:
        if not all(dobra in dobras_cutaneas for dobra in dobras_requeridas):
            faltantes = [dobra for dobra in dobras_requeridas if dobra not in dobras_cutaneas]
            raise KeyError(f"Faltam as seguintes medidas: {', '.join(faltantes)}")
        
        soma_dobras = sum(dobras_cutaneas[dobra] for dobra in dobras_requeridas)
        
        percentual_gordura = round(soma_dobras * 0.095 + 3.64, 2)

    except KeyError as e:
        cadastro['Seis_Dobras'] = {"Erro": str(e)}
        return cadastro
    except Exception as e:
        cadastro['Seis_Dobras'] = {"Erro": f"Erro inesperado: {e}"}
        return cadastro

    cadastro['Seis_Dobras'] = {
        'Soma_Dobras': soma_dobras,
        'Percentual_de_Gordura': percentual_gordura,
    }
    return cadastro

Cadastro_Aluno = Seis_Dobras(Cadastro_Aluno)
```
